```python
import numpy as np
import pandas as pd
```

### One-Sheet EXCEL|CSV

```python
interest_rate = pd.read_csv(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\Interest Rate.csv"
)
real_wage = pd.read_csv(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\Real Wage.csv"
)
adjusted_reserves = pd.read_csv(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\Adjusted Reserves.csv"
)
ppic = pd.read_csv(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\PPIC.csv"
)
```

#### Combine One-Sheet EXCEL|CSV

```python
names = [interest_rate, real_wage, adjusted_reserves, ppic]
one_sheet_combined_data = pd.DataFrame(
    {"observation_date": adjusted_reserves["observation_date"]}
)
for i in names:
    one_sheet_combined_data = pd.merge(
        one_sheet_combined_data, i, on="observation_date", how="outer"
    )
one_sheet_combined_data["observation_date"] = (
    pd.to_datetime(one_sheet_combined_data["observation_date"])
    .dt.to_period("Q")
    .astype(str)
)
one_sheet_combined_data.head()
```

### Multi-Sheet EXCEL

#### GDP

```python
gdp = pd.read_excel(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\GDP&Gov Expenditures.xlsx",
    sheet_name="T10105-Q",
    skiprows=7,
)
gdp = gdp.iloc[[0], :].T.iloc[3:,].reset_index()
gdp.columns = ["time", "GDP"]
gdp["GDP"] = gdp["GDP"] * 1000000
gdp.head()
```

#### Private Consumption

```python
private_consumption = pd.read_excel(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\GDP&Gov Expenditures.xlsx",
    sheet_name="T10105-Q",
    skiprows=7,
)
private_consumption = private_consumption.iloc[[1], :].T.iloc[3:,].reset_index()
private_consumption.columns = ["time", "private_consumption"]
private_consumption["private_consumption"] = (
    private_consumption["private_consumption"] * 1000000
)
private_consumption.head()
```

#### Total Government Expenditure

```python
government_expenditure = pd.read_excel(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\GDP&Gov Expenditures.xlsx",
    sheet_name="T10105-Q",
    skiprows=7,
)
government_expenditure = government_expenditure.iloc[[21], :].T.iloc[3:,].reset_index()
government_expenditure.columns = ["time", "government_expenditure"]
government_expenditure["government_expenditure"] = (
    government_expenditure["government_expenditure"] * 1000000
)
government_expenditure.head()
```

#### Total Government Revenue

```python
government_revenue = pd.read_excel(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\Gov Receipts.xlsx",
    sheet_name="T30100-Q",
    skiprows=7,
)
government_current_receipts = government_revenue.iloc[[0], :].T.iloc[3:,].reset_index()
transfers_payments = government_revenue.iloc[[14], :].T.iloc[3:,].reset_index()
interest_payments = government_revenue.iloc[[10], :].T.iloc[3:,].reset_index()


cashe = (
    government_current_receipts.iloc[:, 1]
    - transfers_payments.iloc[:, 1]
    - interest_payments.iloc[:, 1]
)

government_revenue = pd.DataFrame(
    {"time": government_expenditure["time"], "government_revenue": cashe}
)
government_revenue["government_revenue"] = (
    government_revenue["government_revenue"] * 1000000
)
government_revenue.head()
```

#### Private Non-Residential Investment

```python
private_investment = pd.read_excel(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\GDP&Gov Expenditures.xlsx",
    sheet_name="T10105-Q",
    skiprows=7,
)
gross_private_investment = (private_investment.iloc[[6], :]).T.iloc[3:,].reset_index()
private_residential_investment = (
    (private_investment.iloc[[12], :]).T.iloc[3:,].reset_index()
)
cashe = gross_private_investment.iloc[:, 1] - private_residential_investment.iloc[:, 1]
private_non_residential_investment = pd.DataFrame(
    {
        "time": government_expenditure["time"],
        "private_non_residential_investment": cashe,
    }
)
private_non_residential_investment["private_non_residential_investment"] = (
    private_non_residential_investment["private_non_residential_investment"] * 1000000
)
private_non_residential_investment.head()
```

#### GDP Deflator

```python
deflator = pd.read_excel(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\GDP&Gov Expenditures.xlsx",
    sheet_name="T10109-Q",
    skiprows=7,
)
deflator = deflator.iloc[[0], :].T.iloc[3:,].reset_index()
deflator.columns = ["time", "gdp_deflator"]
deflator.head()
```

#### Population

```python
population = pd.read_excel(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\Population.xlsx",
    sheet_name="T20100-Q",
    skiprows=7,
)
population = population.iloc[[42], :].T.iloc[3:,].reset_index()
population.columns = ["time", "Population"]
population["Population"] = population["Population"] * 1000
population.head()
```

#### Combine Multi-Sheet EXCEL

```python
names = [
    gdp,
    private_consumption,
    government_expenditure,
    government_revenue,
    private_non_residential_investment,
    deflator,
    population,
]
multi_sheet_combined_data = pd.DataFrame({"time": gdp["time"]})
for i in names:
    multi_sheet_combined_data = pd.merge(
        multi_sheet_combined_data, i, on="time", how="outer"
    )
multi_sheet_combined_data.head()
```

#### Combine ALL the data

```python
mydata = pd.merge(
    one_sheet_combined_data,
    multi_sheet_combined_data,
    left_on="observation_date",
    right_on="time",
    how="outer",
)
mydata.rename(
    columns={
        "FEDFUNDS": "Interest Rate",
        "COMPRNFB": "Real Wages",
        "ADJRESSL": "Adjusted Reserves",
        "PPICRM": "PPIC",
    },
    inplace=True,
)
mydata.drop(columns=["observation_date"], inplace=True)
mydata.set_index("time", inplace=True)
mydata.dropna(inplace=True)
mydata.head()
```

#### Save as CSV

```python
mydata.to_csv(
    r"D:\Program Files (x86)\Project\PythonProject\Python Time Series\Replicate\mydata.csv",
    index=True,
)
```
